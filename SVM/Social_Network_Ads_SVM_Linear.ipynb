```python
X = Data[['Age' , 'EstimatedSalary']]
Y = Data['Purchased']
utils.plot_points(X,Y)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X , Y , test_size = 0.25 , random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
S = StandardScaler()
x_train_scaled = S.fit_transform(x_train)
x_test_scaled = S.transform(x_test)
```

```python
print(x_train_scaled)
```

```python
print(x_test_scaled)
```

```python
x_test
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(x_train_scaled , y_train)
```

```python
print(model.predict(S.transform([[30 , 87000]])))
```

```python
print(model.score(x_test_scaled , y_test))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_decision_boundary(X, y, model):
    if isinstance(X, pd.DataFrame):
        X = X.values
    if isinstance(y, pd.Series):
        y = y.values

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))

    grid = np.c_[xx.ravel(), yy.ravel()]
    Z = model.predict(grid)
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
                s=100, facecolors='none', edgecolors='k', linewidths=1.5, label='Support Vectors')
    plt.title('SVM Decision Boundary (Linear Kernel) fot training...')
    plt.legend()
    plt.show()

plot_decision_boundary(x_train_scaled, y_train, model)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

y_prediction = model.predict(x_test_scaled)
cm = confusion_matrix(y_test , y_prediction)

display = ConfusionMatrixDisplay(confusion_matrix = cm , display_labels = model.classes_)
display.plot()
```

```python
from sklearn.metrics import f1_score,recall_score,precision_score

print(f1_score(y_test , y_prediction))
print(recall_score(y_test , y_prediction))
print(precision_score(y_test , y_prediction))
```
